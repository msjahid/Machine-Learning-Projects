<h2><b><span style="color:#009b2d">üíª Step 2.4 | </span><span style="color:#9b006e">üìè Statistical Data Summary</span></b></h2>

```python
def calculate_statistics(df, opacity=0.5, font_size='17px'):
    # Filter numeric columns
    numeric_df = df.select_dtypes(include=['number'])
    
    stat = pd.DataFrame()  # Initialize an empty dataframe

    # Measure of central tendency
    stat['Mean'] = numeric_df.mean()  # for mean of all columns
    stat['Median'] = numeric_df.median()  # for median of all columns
    stat['Mode'] = numeric_df.mode().iloc[0]  # for mode of all columns

    # Measure of dispersion
    stat['Range'] = (numeric_df.max() - numeric_df.min())
    stat['Stdev'] = numeric_df.std()
    stat['Variance'] = numeric_df.var()

    # Interquartile range
    stat['IQR'] = [stats.iqr(numeric_df[i]) for i in numeric_df.columns]

    # Quantiles
    stat['25%'] = numeric_df.quantile(q=0.25, axis=0)
    stat['50%'] = numeric_df.quantile(q=0.50, axis=0)
    stat['75%'] = numeric_df.quantile(q=0.75, axis=0)

    # Minimum and maximum
    stat['Min'] = numeric_df.min()
    stat['Max'] = numeric_df.max()

    # Skewness and Kurtosis
    stat['Skewness'] = numeric_df.skew()
    stat['Kurtosis'] = numeric_df.kurt()

    # Apply the float format
    stat = stat.apply(lambda x: x.map('{:,.2f}'.format))

    # Apply the background color with transparency
    styled_df = stat.style.apply(lambda x: [f'background: rgba(173, 216, 230, {opacity})' for val in x], axis=1)
    
    # Apply font size
    styled_df = styled_df.set_properties(**{'font-size': font_size})
    
    return styled_df
```

```python
# Example usage:
# Assuming you have a DataFrame called 'df' and want to set transparency to 0.5
stats_df = calculate_statistics(df, opacity=0.4, font_size='10px')
stats_df
```

<h2><b><span style="color:#009b2d">üíª Step 3.1 | </span><span style="color:#9b006e"> üìã Columns formating</span></b></h2>

```python
# Define a function to clean column names
def format_column_name(df):
    # Convert column names to lowercase
    df.columns = df.columns.str.lower()
    
    # Replace spaces with underscores
    df.columns = df.columns.str.replace(' ', '_')
    
    # Remove any special characters except underscores
    df.columns = df.columns.map(lambda x: re.sub(r'\W+', '', x))
    
    return df
```

```python
# Rename columns to lowercase and replace spaces with underscores
renamed_columns = pd.DataFrame({
    'Previous Name': [color.RED + col + color.RESET for col in df.columns],
    'New Name': [color.GREEN + re.sub(r'\W+', '', col.lower().replace(' ', '_')) + color.RESET for col in df.columns]
})

# Pretty print using tabulate for better formatting
print(tabulate(renamed_columns, headers='keys', tablefmt='grid'))

# Apply the function to clean column names
df = format_column_name(df)
```

```python
def apply_duplicated_styling(df):
    
    # Define a function to apply font size and color to values
    def apply_style(value):
        if value:
            return 'font-size: 20px; color: red'
        else:
            return 'font-size: 20px; color: green'

    # Define a function to apply font size to column names
    def apply_column_name_style(value):
        return 'font-size: 20px'

    # Transpose the DataFrame and check for duplicated values along the rows
    column_duplicates = df.T.duplicated()

    # Apply styling to the DataFrame containing duplicated values
    styled_column_duplicates = column_duplicates.to_frame().style.applymap(apply_style).background_gradient(axis=0)

    # Apply styling to the column names
    styled_column_duplicates.set_table_styles([{'selector': 'th', 'props': [('font-size', '20px')]}])

    return styled_column_duplicates
```

```python
def variable_summary(df, opacity=0.5, font_size='20px'):
    # Initialize an empty list to store individual summaries
    summaries = []

    # Iterate through each column in the DataFrame
    for col in df.columns:
        dtype = df[col].dtype  # Data type of the column
        count = df[col].count()  # Count of non-null values
        unique = df[col].nunique()  # Number of unique values
        missing = df[col].isna().sum()  # Number of missing values

        # Append the summary information to the list
        summaries.append([col, dtype, count, unique, missing])

    # Create a DataFrame from the list of summaries
    summary_df = pd.DataFrame(summaries, columns=['Variable', 'Dtype', 'Count', 'Unique', 'Missing'])

    # Apply background color with opacity
    styled_summary_df = summary_df.style.apply(lambda x: [f'background: rgba(42, 46, 120, {opacity})' for val in x], axis=1)

    # Apply font size to the entire DataFrame
    styled_summary_df.set_table_styles([{'selector': '', 'props': [('font-size', font_size)]}])

    return styled_summary_df
```

  <br><br>
  <strong><u>Multivariate Analysis:</u></strong> Multivariate analysis involves analyzing more than two variables simultaneously. It helps to understand the relationships and interactions between multiple variables. Common techniques include principal component analysis (PCA), factor analysis, and cluster analysis.
  <br><br>

```python
def generate_wordcloud(df, column):

    # Concatenate all the text data from the specified column
    text_data = ' '.join(df[column].dropna())

    # Create the word cloud object with a gray background
    wordcloud = WordCloud(width=800, height=400, background_color='gray').generate(text_data)

    # Create a centered figure
    plt.figure(figsize=(10, 6))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')

    # Show the word cloud
    plt.show()
```

```python
# Show the brand
generate_wordcloud(df, column='brand')
```

```python
def plot_brand_distribution(df, column_name, subplot_color='gray'):

    # Calculate the counts of each brand
    brand_counts = df['brand'].value_counts()

    # Define a custom color palette
    custom_palette = sns.color_palette("deep", len(brand_counts))

    # Plotting
    fig, axes = plt.subplots(1, 2, figsize=(14, 6), facecolor=subplot_color)

    # Countplot for the distribution of brands
    sns.countplot(data=df, x='brand', ax=axes[0], palette=custom_palette)
    axes[0].set_xlabel('Brand')
    axes[0].set_ylabel('Count')
    axes[0].set_facecolor(subplot_color)  # Set subplot background color

    # Pie chart for the distribution of brands
    axes[1].pie(brand_counts.values, labels=brand_counts.index, autopct='%1.2f%%', colors=custom_palette)
    axes[1].set_facecolor(subplot_color)  # Set subplot background color

    # Set titles for the brand plots
    axes[0].set_title('Distribution of Brands')
    axes[1].set_title('Distribution of Brands')

    # Set the super title for the entire figure
    plt.suptitle('Analysis of Brand Distribution', fontsize=16, color='white')

    # Adjust layout
    plt.tight_layout()

    # Show the plot
    plt.show()
```

```python
plot_brand_distribution(df, column_name='brand')
```

<div class="alert alert-block alert-info">
    <center style="font-size: 22px; color: Blue;">_‚ô•_‚ô•_‚ô• <u>  Data Distributions Explained </u> ‚ô•_‚ô•_‚ô•_</center>
    <br><br>
<p style="font-family: monospace; font-size: 18px; color: #CD7A31;">
  <strong>In statistics</strong>, a <strong>distribution</strong> refers to the pattern or spread of values that a variable can take on within a dataset. It describes how the values of a variable are distributed or spread out across different possible outcomes.
</p>
<p style="font-family: monospace; font-size: 16px; color: #CD7A31;">
    <strong><u>Normal Distribution (Gaussian Distribution):</u></strong> A symmetric bell-shaped distribution where the mean, median, and mode are all equal. It is characterized by its mean and standard deviation.
    <br><br>
    <strong><u>Uniform Distribution:</u></strong> A distribution where all values within a specified range have an equal probability of occurring. It is characterized by a constant probability density function over the range.
    <br><br>
    <strong><u>Binomial Distribution:</u></strong> A discrete probability distribution that describes the number of successes in a fixed number of independent Bernoulli trials, each with the same probability of success.
    <br><br>
    <strong><u>Poisson Distribution:</u></strong> A discrete probability distribution that describes the number of events occurring in a fixed interval of time or space, given a constant rate of occurrence and independence between events.
    <br><br>
    <strong><u>Exponential Distribution:</u></strong> A continuous probability distribution that describes the time between events in a Poisson process, where events occur continuously and independently at a constant rate.
    <br><br>
    <strong><u>Gamma Distribution:</u></strong> A continuous probability distribution that generalizes the exponential distribution and describes the time until the nth event in a Poisson process.
    <br><br>
    <strong><u>Chi-Square Distribution:</u></strong> A continuous probability distribution that arises in the context of hypothesis testing and describes the distribution of the sum of squared standard normal deviates.
    <br><br>
    <strong><u>Student's t-Distribution:</u></strong> A continuous probability distribution that is similar in shape to the normal distribution but has heavier tails. It is commonly used in hypothesis testing when the sample size is small and the population standard deviation is unknown.
    <br><br>

  Left skew and right skew are terms used to describe the shape of the distribution of data points in a histogram or probability density plot.
  <br><br>
  <strong><u>Left Skew (Negative Skew):</u></strong> A distribution is left-skewed if the left tail (the side with smaller values) is longer or stretched out compared to the right tail. This means that the majority of the data points are concentrated on the right side of the distribution, while a few extreme values pull the mean towards the left. The mean is typically less than the median in a left-skewed distribution.
  <br><br>
  <strong><u>Right Skew (Positive Skew):</u></strong> A distribution is right-skewed if the right tail (the side with larger values) is longer or stretched out compared to the left tail. This means that the majority of the data points are concentrated on the left side of the distribution, while a few extreme values pull the mean towards the right. The mean is typically greater than the median in a right-skewed distribution.

```python
def anova_comparison(df, columns=None, target_variable=None):
    try:
        if columns is None:
            raise ValueError("Columns argument is required.")
            
        if target_variable is None:
            raise ValueError("Group column argument is required.")
            
        if target_variable not in df.columns:
            raise ValueError(f"Group column '{target_variable}' not found in DataFrame.")
            
    except ValueError as ve:
        print(ve)
        return

    alpha = 0.05

    # Lists to store results
    comparison_results = []
    columns_list = []
    category1_list = []
    category2_list = []
    
    formatted_columns = [re.sub(r'[_\.]', ' ', col).title() for col in columns]

    # Perform ANOVA for each column
    for column in columns:
        groups = df[target_variable].unique()
        group_pairs = list(combinations(groups, 2))  # Generate all possible pairs of groups
        
        for category1, category2 in group_pairs:
            data1 = df[df[target_variable] == category1][column]
            data2 = df[df[target_variable] == category2][column]
        
            f_stat, p_value = f_oneway(data1, data2)

            if abs(p_value) <= alpha:
                result = color.GREEN + 'Yes' + color.RESET
            else:
                result = color.RED + 'No' + color.RESET

            comparison_results.append(result)
            columns_list.append(column)
            category1_list.append(category1)
            category2_list.append(category2)

    # Create DataFrame
    df_comparison = pd.DataFrame({
        'Group': columns_list,
        'Category->1': category1_list,
        'Category->2': category2_list,
        'Significant Difference': comparison_results
    })
    
#     print(f"{color.BLUE}ANOVA comparison for {name}".center(80) + color.RESET)

    # Loop through each column and print the corresponding table
    for i, column in enumerate(columns):
        # Print table for the current column with colored column names
        print(f"{color.BLUE}ANOVA test for {(formatted_columns[i])}".center(80) + color.END)
        print(tabulate(df_comparison[df_comparison['Group'] == column], headers='keys', tablefmt="grid"))
        # Print an empty line for separation
        print()
```

```python
# Split the data into features (X) and target variable (y)
X = df.drop(["price"], axis=1)  # Drop "price"
y = df["price"].values
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
# One-hot encode the categorical features in the training and testing sets
X_train_encoded = pd.get_dummies(X_train)
X_test_encoded = pd.get_dummies(X_test)
```

<a id="train"></a>
<h2><b><span style="color:#009b2d">üíª Step 5.3   </span><span style="color:#9b006e">| ‚öôÔ∏è Model Training</span></b></h2>

```python
def knn_performance(X_train_encoded, y_train, X_test_encoded):
    k_values = []
    accuracies = []
    print(f"{color.BLUE}{color.BOLD}KNN Performance with Different Values Of k With Accuracy".center(80) + color.END + "\n")
    for i in range(1, 50):
        model = KNeighborsRegressor(n_neighbors = i)
        model.fit(X_train_encoded, y_train)  # fit the model
        prediction = model.predict(X_test_encoded)  # make prediction on test set
        accuracy = r2_score(y_test, prediction)
        # Printing in red color
        print(color.RED + f"For k={i}" + color.END)
        # Printing accuracy in green color
        print(color.GREEN + f"Accuracy is - {accuracy*100:.2f}%" + color.END)
        k_values.append(i)
        accuracies.append(accuracy)
    return k_values, accuracies
```

```python
# Call the knn_performance function
k_values, accuracies = knn_performance(X_train_encoded, y_train, X_test_encoded)
```

```python
def create_custom_plot(k, accu):
    # Set seaborn style and context
    sns.set(style="whitegrid")
    sns.set_context("notebook", font_scale=1.2)

    # Create the plot with gray facecolor and subplot color, and specify the figure size
    plt.figure(facecolor='gray', figsize=(15, 6))
    ax = plt.subplot(facecolor='gray')

    # Plot the data
    sns.lineplot(x=k, y=accu, color='skyblue', linewidth=2.5, marker='o', markersize=8)

    # Add labels and title
    plt.xlabel("Value of K", fontsize=14, fontweight='bold', color='white')
    plt.ylabel("R2 Score", fontsize=14, fontweight='bold', color='white')
    plt.title("R2 Score vs. Value of K", fontsize=16, fontweight='bold', color='white')

    # Add grid lines
    plt.grid(True, linestyle='--', alpha=0.7)

    # Customize tick labels
    plt.xticks(fontsize=12, color='white')
    plt.yticks(fontsize=12, color='white')

    # Show plot
    plt.show()
```

```python
# Define parameters to test in KNN regression
parameters = {
    'n_neighbors': [3, 5, 7, 9, 11],  # Example values for the number of neighbors
    'weights': ['uniform', 'distance'],  # Weighting strategies for neighbors
    'metric': ['euclidean', 'manhattan']  # Distance metrics
}

# Create an instance of KNeighborsRegressor
knn = KNeighborsRegressor()

# Perform grid search using cross-validation
knn_cv = GridSearchCV(estimator=knn, param_grid=parameters, cv=10).fit(X_train_encoded, y_train)

# Get the best score, k values and hyperparameters from GridSearchCV
best_score = knn_cv.best_score_
best_params = knn_cv.best_params_
best_n_neighbors = best_params['n_neighbors']

# Print the best hyperparameters, k values and accuracy
print(f"Tuned hyperparameters: {color.CYAN}{best_params}{color.END}")
print(f"The best value of k (n_neighbors) is: {color.BLUE}{best_n_neighbors}{color.END}")
print(f"Accuracy is - {color.GREEN}{best_score*100:.2f}%{color.END}")
```

```python
def predict_price(model, X_train_encoded, user_data):
    # Create a DataFrame from the user data
    user_df = pd.DataFrame([user_data], columns=['brand', 'processor_speed', 'ram_size', 'storage_capacity', 'screen_size', 'weight'])

    # Scale the numerical features using the provided scaler object
    numerical_features = ['processor_speed', 'ram_size', 'storage_capacity', 'screen_size', 'weight']
    user_df[numerical_features] = scaler.transform(user_df[numerical_features])

    # One-hot encode the 'brand' column
    user_df_encoded = pd.get_dummies(user_df)

    # Ensure the encoded DataFrame has the same columns as the training data
    missing_cols = set(X_train_encoded) - set(user_df_encoded.columns)
    for col in missing_cols:
        user_df_encoded[col] = 0

    user_df_encoded = user_df_encoded[X_train_encoded.columns]

    # Use the trained model to predict the price for the user data
    predicted_price = model.predict(user_df_encoded)

    # Round the predicted price to two decimal places
    rounded_price = "{:.2f}".format(predicted_price[0])

    return rounded_price
```
