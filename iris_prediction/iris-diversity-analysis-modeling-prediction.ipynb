<h2><b><span style="color:#009b2d">üå∏Step 2.4 | </span><span style="color:#9b006e">üìè Statistical Data Summary</span></b></h2>

```python
def calculate_statistics(df, opacity=0.5, font_size='17px'):
    # Filter numeric columns
    numeric_df = df.select_dtypes(include=['number'])
    
    stat = pd.DataFrame()  # Initialize an empty dataframe

    # Measure of central tendency
    stat['Mean'] = numeric_df.mean()  # for mean of all columns
    stat['Median'] = numeric_df.median()  # for median of all columns
    stat['Mode'] = numeric_df.mode().iloc[0]  # for mode of all columns

    # Measure of dispersion
    stat['Range'] = (numeric_df.max() - numeric_df.min())
    stat['Stdev'] = numeric_df.std()
    stat['Variance'] = numeric_df.var()

    # Interquartile range
    stat['IQR'] = [stats.iqr(numeric_df[i]) for i in numeric_df.columns]

    # Quantiles
    stat['25%'] = numeric_df.quantile(q=0.25, axis=0)
    stat['50%'] = numeric_df.quantile(q=0.50, axis=0)
    stat['75%'] = numeric_df.quantile(q=0.75, axis=0)

    # Minimum and maximum
    stat['Min'] = numeric_df.min()
    stat['Max'] = numeric_df.max()

    # Skewness and Kurtosis
    stat['Skewness'] = numeric_df.skew()
    stat['Kurtosis'] = numeric_df.kurt()

    # Apply the float format
    stat = stat.apply(lambda x: x.map('{:,.2f}'.format))

    # Apply the background color with transparency
    styled_df = stat.style.apply(lambda x: [f'background: rgba(173, 216, 230, {opacity})' for val in x], axis=1)
    
    # Apply font size
    styled_df = styled_df.set_properties(**{'font-size': font_size})
    
    return styled_df
```

```python
# Assuming you have a DataFrame called 'df' and want to set transparency to 0.5
# Drop the ID column from the DataFrame
df_without_id = df.drop(columns=['Id'])

# Calculate statistics with the modified DataFrame
stats_df = calculate_statistics(df_without_id, opacity=0.4, font_size='12px')

# Display the resulting DataFrame
stats_df
```

<h2><b><span style="color:#009b2d">üå∏ Step 3.1 | </span><span style="color:#9b006e"> üìã Columns formating</span></b></h2>

```python
def format_column_name(df):
    # Define a function to format a single column name
    def format_single_column_name(column_name):
        
        # Remove "Cm" suffix if present
        if column_name[-2:] == "Cm":
            column_name = column_name[:-2].strip()  # Use strip to remove whitespace
            
        # Replace camel case with underscores
        column_name = re.sub(r'(?<!^)(?=[A-Z])', '_', column_name)
        
        # Convert to lowercase
        column_name = column_name.lower()
        
        return column_name  # Return the modified column name
    
    # Apply format_single_column_name function to each column name
    mapped_names = [format_single_column_name(col) for col in df.columns]
    
    # Create a dictionary mapping original column names to modified column names
    column_mapping = dict(zip(df.columns, mapped_names))
    
    # Rename columns using the mapping
    df = df.rename(columns=column_mapping)
    
    return df
```

```python
previous_name = df.copy()

# Apply the function to clean column names
df = format_column_name(df)

# Rename columns to lowercase and replace spaces with underscores
renamed_columns = pd.DataFrame({
    'Previous Name': [color.RED + col + color.RESET for col in previous_name.columns],
    'New Name': [color.GREEN + col + color.RESET for col in df.columns]
})

# Pretty print using tabulate for better formatting
print(tabulate(renamed_columns, headers='keys', tablefmt='grid'))

# Apply the function to clean column names
df = format_column_name(df)
```

```python
def plot_missing_values_heatmap(df):
    # Calculate the sum of missing values excluding the first column
    missing_values_sum = df.isna().iloc[:, 1:].sum()

    # Create a heatmap plot
    plt.figure(figsize=(22, 4), facecolor='gray')
    ax = sns.heatmap(missing_values_sum.to_frame(name='').T.rename(columns=lambda x: x.replace('_', ' ').title()), 
                     cmap=sns.diverging_palette(250, 30, l=65, center="dark", as_cmap=True), annot=True, fmt='0.0f',
                     annot_kws={"fontsize":30, "fontweight": "bold"},  # Adjust font size and weight for annotations
                     linecolor='grey', linewidth=1)

    # Set title and adjust fontsize
    plt.title('Count of Missing Values', fontsize=18, fontweight='bold')

    # Adjust font size and weight of x-axis and y-axis labels
    ax.set_xticklabels([label.replace('_', ' ').title() for label in missing_values_sum.index], fontsize=15, fontweight='bold')
    ax.set_yticklabels(ax.get_yticklabels(), fontsize=15, fontweight='bold')

    # Show the plot
    plt.show()
```

```python
def apply_duplicated_styling(df):
    
    # Define a function to apply font size and color to values
    def apply_style(value):
        if value:
            return 'font-size: 20px; color: red'
        else:
            return 'font-size: 20px; color: green'

    # Define a function to apply font size to column names
    def apply_column_name_style(value):
        return 'font-size: 20px'

    # Transpose the DataFrame and check for duplicated values along the rows
    column_duplicates = df.T.duplicated()

    # Apply styling to the DataFrame containing duplicated values
    styled_column_duplicates = column_duplicates.to_frame().style.applymap(apply_style).background_gradient(axis=0)

    # Apply styling to the column names
    styled_column_duplicates.set_table_styles([{'selector': 'th', 'props': [('font-size', '20px')]}])

    return styled_column_duplicates
```

```python
def variable_summary(df, opacity=0.5, font_size='20px'):
    # Initialize an empty list to store individual summaries
    summaries = []

    # Iterate through each column in the DataFrame
    for col in df.columns:
        dtype = df[col].dtype  # Data type of the column
        count = df[col].count()  # Count of non-null values
        unique = df[col].nunique()  # Number of unique values
        missing = df[col].isna().sum()  # Number of missing values

        # Append the summary information to the list
        summaries.append([col, dtype, count, unique, missing])

    # Create a DataFrame from the list of summaries
    summary_df = pd.DataFrame(summaries, columns=['Variable', 'Dtype', 'Count', 'Unique', 'Missing'])

    # Apply background color with opacity
    styled_summary_df = summary_df.style.apply(lambda x: [f'background: rgba(42, 46, 120, {opacity})' for val in x], axis=1)

    # Apply font size to the entire DataFrame
    styled_summary_df.set_table_styles([{'selector': '', 'props': [('font-size', font_size)]}])

    return styled_summary_df
```

<h2><b><span style="color:#009b2d">üå∏ Step 4.1 | </span><span style="color:#9b006e"> üìà Individual Variables Analysis</span></b></h2>

```python
def plot_brand_distribution(df, column_name, subplot_color='silver'):
    
    # Set the seaborn style to remove white grid lines
    sns.set_style("dark")

    # Calculate the counts of each brand
    counts = df[column_name].value_counts()

    # Define a custom color palette
    custom_palette = sns.color_palette("deep", len(counts))

    # Plotting
    fig, axes = plt.subplots(1, 2, figsize=(14, 6), facecolor=subplot_color)

    # Countplot for the distribution of brands
    sns.countplot(data=df, x=column_name, ax=axes[0], palette=custom_palette)
    
    # Get the formatted column name
    formatted_column_name = column_name.title()

    # Set the x-axis label for the first subplot
    axes[0].set_xlabel(formatted_column_name)
    axes[0].set_ylabel('Count')
    axes[0].set_facecolor(subplot_color)  # Set subplot background color

    # Pie chart for the distribution of brands
    axes[1].pie(counts.values, labels=counts.index, autopct='%1.2f%%', colors=custom_palette)
    axes[1].set_facecolor(subplot_color)  # Set subplot background color

    # Set titles for the brand plots
    axes[0].set_title(f'Distribution of {formatted_column_name}')
    axes[1].set_title(f'Distribution of {formatted_column_name}')

    # Set the super title for the entire figure
    plt.suptitle(f'Analysis of {formatted_column_name} Distribution', fontsize=16, color='black')

    # Adjust layout
    plt.tight_layout()

    # Show the plot
    plt.show()
```

```python
def plot_correlation_heatmap(df, columns):
    
    palette = sns.color_palette("dark", len(columns))
    
    # Calculate the correlation matrix
    corrmat = df.corr()

    # Define the colormap
    cmap = sns.diverging_palette(240, 10, s=75, l=40, n=9, center="dark", as_cmap=True)

    # Create a figure and axes
    f, ax = plt.subplots(figsize=(20, 12), facecolor='silver')

    # Plot the heatmap with specified colors
    heatmap = sns.heatmap(corrmat, annot=True, cmap=cmap, fmt=".2f", vmin=-1, vmax=1, annot_kws={"size": 14}, square=True)

    # Set font size for x-axis and y-axis labels
    heatmap.set_xticklabels([col.replace('_', ' ').title() for col in corrmat.columns], rotation=45, fontsize=16)
    heatmap.set_yticklabels([col.replace('_', ' ').title() for col in corrmat.index], rotation=0, fontsize=16)
    
    # Modify tick label color using the palette
    for i, label in enumerate(ax.get_xticklabels()):
        label.set_color(palette[i])
    for i, label in enumerate(ax.get_yticklabels()):
        label.set_color(palette[i])

    # Add title
    plt.title('Correlation Heat Map of Different Flowers Features', pad=45, fontsize=25, color='black')

    # Show plot
    plt.show()
```

```python
def plot_violin_boxplots_grid(df, columns, columns_per_row=2):
    # Calculate the number of rows required based on the number of columns and columns per row
    num_cols = len(columns)
    num_rows = (num_cols - 1) // columns_per_row + 1
    
    # Create a figure and subplots
    fig, axes = plt.subplots(num_rows, columns_per_row, figsize=(14, 4.5 * num_rows), facecolor='silver')

    # Loop through the columns and create violin plots with box plots inside the subplots
    for i, col in enumerate(columns):
        row_index = i // columns_per_row
        col_index = i % columns_per_row

        # Plot violin plot
        sns.violinplot(x='species', y=col, data=df, ax=axes[row_index, col_index])

        # Plot boxplot with adjusted width
        sns.boxplot(x='species', y=col, data=df, ax=axes[row_index, col_index], width=0.1, color="black", linewidth=2.5, palette='Set2')

        # Set title and labels
        axes[row_index, col_index].set_title(col.replace('_', ' ').title(), fontsize=16)
        axes[row_index, col_index].set_xlabel('')
        axes[row_index, col_index].set_ylabel('')
        # Set subplot background color
        axes[row_index, col_index].set_facecolor('silver')

    plt.tight_layout()
    plt.show()
```

```python
def create_custom_pairplot(df, columns, target):
    # Set seaborn theme and color palette
    sns.set_theme(style="whitegrid")
    palette = sns.color_palette("dark", len(columns))
    sns.set_context(font_scale=1.5)
    tick_label_palette = sns.color_palette("Set3", len(df))
    
    # Create a gray background figure
    plt.figure(figsize=(10, 8), facecolor='silver')
    
    # Create pairplot with gray background
    pairplot = sns.pairplot(df, hue=target, corner=True, palette='rocket')
    pairplot.fig.set_facecolor('silver')
    for ax in pairplot.axes.flat:
        if ax is not None:
            ax.set_facecolor('silver')
    
    # Set x-axis titles with modified column names
    for i, ax in enumerate(pairplot.axes[-1,:]):
        title = df.columns[i].replace('_', ' ').title()
        ax.set_xlabel(title, fontsize=14, color=palette[i], fontweight='bold')
    
    # Set y-axis titles with modified column names
    for i, ax in enumerate(pairplot.axes[:,0]):
        title = df.columns[i].replace('_', ' ').title()
        ax.set_ylabel(title, fontsize=14, color=palette[i], fontweight='bold')
    
    # Change legend title from 'brand' to 'Brand'
    pairplot._legend.set_title(target.title())
    plt.show()
```

```python
def z_test_comparison(df, columns=None, unique_col=None, name=None):
    try:
        if columns is None:
            raise ValueError("Columns argument is required.")
            
        if unique_col is None:
            raise ValueError("Unique column argument is required.")
            
        if unique_col not in df.columns:
            raise ValueError(f"Unique column '{unique_col}' not found in DataFrame.")
            
    except ValueError as ve:
        print(ve)
        return

    # Get the unique varieties
    specific = df[unique_col].unique()
    
    alpha=0.05

    # Lists to store results
    comparison_results = []
    columns_list = []
    category1_list = []
    category2_list = []

    # Perform Z-test for each pair of varieties and each column
    for column in columns:
        for category1, category2 in combinations(specific, 2):
            data1 = df[df[unique_col] == category1][column]
            data2 = df[df[unique_col] == category2][column]

            z_score, p_value = wstat.ztest(data1, data2)

            if abs(p_value) <= alpha:
                result = color.GREEN + 'Yes' + color.RESET
            else:
                result = color.RED + 'No' + color.RESET

            comparison_results.append(result)
            columns_list.append(column)
            category1_list.append(category1)
            category2_list.append(category2)

    # Create DataFrame
    df_comparison = pd.DataFrame({
        'Group': columns_list,
        'Category->1': category1_list,
        'Category->2': category2_list,
        'Significant Difference': comparison_results
    })
    
    print(f"{color.BLUE}Z-test for variety of {name}".center(80) + color.END)

    # Loop through each column and print the corresponding table
    for column in columns:
        # Print table for the current column with colored column names
        print(tabulate(df_comparison[df_comparison['Group'] == column], headers='keys', tablefmt="double_grid"))
        # Print an empty line for separation
        print()
```

```python
def clf_plot(y_pred):
    '''
    1) Plotting Confusion Matrix
    2) Plotting Classification Report'''
    # Define the colormap
    cmap = sns.diverging_palette(240, 10, s=75, l=40, n=9, center="dark", as_cmap=True)
    
    cm = metrics.confusion_matrix(y_test, y_pred)
    cr = pd.DataFrame(metrics.classification_report(y_test, y_pred, digits=3, output_dict=True)).T
    cr.drop(columns='support', inplace=True)
    
    fig, ax = plt.subplots(1, 2, figsize=(15, 5), facecolor='silver')
    
    # Left AX : Confusion Matrix
    ax[0] = sns.heatmap(cm, annot=True, fmt='', cmap=cmap, ax=ax[0])
    ax[0].set_xlabel('Predicted labels', fontsize=18)
    ax[0].set_ylabel('True labels', fontsize=18)
    ax[0].set_title('Confusion Matrix', fontsize=25)
    ax[0].xaxis.set_ticklabels(['setosa', 'versicolor', 'virginica'])
    ax[0].yaxis.set_ticklabels(['setosa', 'versicolor', 'virginica'])
    
    plt.gca().set_facecolor('silver')  # Set subplot background color to silver
    
    # Right AX : Classification Report
    ax[1] = sns.heatmap(cr, cmap=cmap, annot=True, linecolor='white', linewidths=0.5, ax=ax[1])
    ax[1].xaxis.tick_top()
    ax[1].set_title('Classification Report', fontsize=25)
    plt.show()
```
